## Construindo o modelo preditivo

```python
from prophet import Prophet
```

```python
#Para utilizar o prophet, ele utiliza apenas duas colunas que precisam ser chamadas de 'ds' contendo a data, e 'y' o que queremos prever.
df_prophet = df_data[['data', 'contagem']].rename(columns={'data':'ds', 'contagem':'y'})
df_prophet.head()
```

```python
df_prophet = df_prophet.groupby('ds')['y'].sum().reset_index()
df_prophet.head()
```

```python
#Criando semente aleatória
import numpy as np

np.random.seed(4587)
```

```python
#Criando e treinando o modelo
modelo = Prophet()
modelo.fit(df_prophet)

#Criando dataframe de predição para 90 dias
futuro = modelo.make_future_dataframe(periods=90, freq='D')
previsao = modelo.predict(futuro)
```

```python
fig1 = modelo.plot(previsao)
```

Esse gráfico representa a nossa previsão onde:

-   Pontos pretos: Valores reais do dataframe
-   Linha azul escura: É a previsão do modelo
-   Linha azul clara: É o intervalo de confiança

O ideal é que os pontos pretos fiquem dentro do intervalo de confiança

```python
#Verificando o dataframe de previsão geradoa partir do futuro. A coluna 'y_hat' é o valor que o modelo está prevendo, lower e upper são o invervalo de confiança superior e inferior
previsao[['ds','yhat', 'yhat_lower', 'yhat_upper']]
```

```python
fig2 = modelo.plot_components(previsao)
```

Esses componentes plotados pelo fig2 é o que o modelo do prophet utilizou para prever os dados

## Separando Modelo em Treino e Teste

De primeiro momento utilizamos todos os dados para prever dentro do modelo, mas o ideal é separa-los entre treino e teste.


```python
df_prophet.shape
```

```python
#Criando df de treino com 80% dos dados
df_treino = pd.DataFrame()
df_treino['ds'] = df_prophet['ds'][:584]
df_treino['y'] = df_prophet['y'][:584]
```

```python
#Criando df de teste com 20% dos dados
df_teste = pd.DataFrame()
df_teste['ds'] = df_prophet['ds'][584:]
df_teste['y'] = df_prophet['y'][584:]
```

```python
np.random.seed(4587)

modelo = Prophet()
modelo.fit(df_treino)
futuro = modelo.make_future_dataframe(periods=150, freq='D')
previsao = modelo.predict(futuro)
```

```python
fig1 = modelo.plot(previsao)
plt.plot(df_teste['ds'], df_teste['y'], '.r')
```

```python
fig2 = modelo.plot_components(previsao)
```

Quando separamos o modelo em 2 bases, de teste e treino, o prophet dessa vez não conseguiu notar a sazonalidade anual,
ele entendeu que os dados estavam com uma tendencia de subir, e não que após o encerramento de uma no o padrão era os dados se repetir e descer

## Obtendo Metricas

```python
#Criando data frame para comparar o valor previsto (yhat) com os dados reais no df de teste (y)
df_previsao = previsao[['ds','yhat']]
df_comparacao = pd.merge(df_previsao, df_teste, on='ds')
df_comparacao
```

Para calcularmos o quanto o modelo está errando, faremos a seguinte formula:

MSE = Soma(y - yhat)² / n

n = numero de linhas do df

obs: O MSE trará um valor muito alto pois está elevando ao quadrado, tanto para retirar valores negativos quanto para punir valores errados.
Para tratarmos isso faremos a raiz quadrada desse MSE, tendo um RMSE

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(df_comparacao['y'], df_comparacao['yhat'])

rmse = np.sqrt(mse)

print(f'MSE: {mse}, RMSE: {rmse}')
```

Nosso modelo está errando 21471 bicicletas +/-

## Melhorando as previsões

```python
np.random.seed(4587)

#Adicionando a sazonalidade anual
modelo = Prophet(yearly_seasonality=True)
modelo.fit(df_treino)
futuro = modelo.make_future_dataframe(periods=150, freq='D')
previsao = modelo.predict(futuro)
```

```python
fig1 = modelo.plot(previsao)
plt.plot(df_teste['ds'], df_teste['y'], '.r')
```

```python
fig2 = modelo.plot_components(previsao)
```

Agora com o modelo entendendo que há sazonalidade anual, o modelo aparenta estar mais assertivo

```python
#obtendo metricas novamente

df_previsao = previsao[['ds','yhat']]
df_comparacao = pd.merge(df_previsao, df_teste, on='ds')
df_comparacao

```

```python
mse = mean_squared_error(df_comparacao['y'], df_comparacao['yhat'])
rmse = np.sqrt(mse)
print(f'MSE: {mse}, RMSE: {rmse}')
```

Agora o nosso modelo está errando 6166 bicicletas, o que é uma melhora de 21 mil que estávamos errando. Mas podemos melhorar 

## Removendo Outliers

```python
#df antes de separarmos para teste e treino
df_prophet
```

```python
#Iremos remover os outliers de acordo com o nosso intervalo de confiança

np.random.seed(4587)

modelo = Prophet()
modelo.fit(df_prophet)
futuro = modelo.make_future_dataframe(periods=0)
previsao = modelo.predict(futuro)
```

```python
#Pegando os valores no intervalo de confiança
sem_outliers = df_prophet[(df_prophet['y'] > previsao['yhat_lower']) & (df_prophet['y'] < previsao['yhat_upper'])]
```

```python
sem_outliers
```

```python
#Criando df de treino e teste

df_treino = pd.DataFrame()
df_treino['ds'] = sem_outliers['ds'][:505]
df_treino['y'] = sem_outliers['y'][:505]
```

```python
df_teste = pd.DataFrame()
df_teste['ds'] = sem_outliers['ds'][505:]
df_teste['y'] = sem_outliers['y'][505:]
```

```python
np.random.seed(4587)
modelo_sem_outliers = Prophet(yearly_seasonality=True)
modelo_sem_outliers.fit(df_treino)
futuro = modelo_sem_outliers.make_future_dataframe(periods=150, freq='D')
previsao = modelo_sem_outliers.predict(futuro)
```

```python
fig1 = modelo_sem_outliers.plot(previsao)
plt.plot(df_teste['ds'], df_teste['y'], '.r')
```

```python
df_previsao = previsao[['ds','yhat']]
df_comparacao = pd.merge(df_previsao, df_teste, on='ds')
df_comparacao
```

```python
mse = mean_squared_error(df_comparacao['y'], df_comparacao['yhat'])
rmse = np.sqrt(mse)
print(f'MSE: {mse}, RMSE: {rmse}')
```

Agora estamos errando apenas 3k de bicicletas o que é muito menor do que 6k e ainda menor do que 21k
