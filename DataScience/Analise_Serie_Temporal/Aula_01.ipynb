Será que o outono e a primavera são iguais estatisticamente falando?

```python
from scipy.stats import mannwhitneyu
```

```python
primavera = df_limpo[df_limpo['estacao'] =='Primavera']['contagem']
outono = df_limpo[df_limpo['estacao'] == 'Outono']['contagem']

u_statistic, p_value = mannwhitneyu(primavera, outono, alternative='two-sided')
u_statistic, p_value
```

Quando o p value é menor que 0.05 descartamos a hipótese nula, e ficamos com a hipótese alternativa, neste caso, que outono e primavera são diferentes estatisticamente falando.
