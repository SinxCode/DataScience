```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Arrumando a Base

```python
df = pd.read_csv('C:/Users/gabrielbrito/DataScienceProjects/DataScience/Dados/imigrantes_canada.csv')
```

```python
df.head()   
```

```python
df.info()
```

```python
#Setando o index como o país
df.set_index('País', inplace=True)
```

```python
#Criando variável para armazenar os nomes das colunas de ano
anos = list(map(str,range(1980, 2014)))
```

```python
anos
```

```python
#Criando uma series somente com os dados do brasil e dos anos
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
#Criando dicionario da series anterior nomeando as colunas de ano e imigrantes
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}

#Transformando o dicionário em um df
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

## Plotagem

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.xticks(['1980', '1985', '1990','1995', '2000', '2005','2010'])
plt.show()
```

## Criando uma Figura

A figura diferente de apenas usar o plot, permite mais funcionalidades e personalizações para o gráfico

```python
#Fig permite que uma figura seja criada (o espaço em branco), o ax representa os eixos
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
#Deixando o eixo X automaticamente em 5 em 5
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
#Setando o titulo em figura
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
#adicionando rotulos na figura
ax.set_xlabel('Ano')
ax.set_ylabel('Numero de Imigrantes')
plt.show()
```

## Plotando mais de um grafico na mesma figura

```python
#Dentro do parametro do subplot o 1 indica que ele tera uma linha só, e o 2 indica que serão 2 gráficos nessa msm linha
fig, axs = plt.subplots(1, 2, figsize=(15,5))
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da Imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de Imigrantes')
axs[1].grid()

plt.show()
```

# Criando subplots em 2 direções

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

#Estrutura de repetição para deixar o eixo X de todos os gráficos de 5 em 5
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de Imigrantes')

plt.show()
```

Os gráficos plotados mostra que os valores no eixo y não estão na mesma escala, alguns o máximo chega a ser 1500, outros 2500 e outros 6000. É necessário ajustar esse valor mínimo e máximo para a análise não ficar enviesada.

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigração dos 4 maiores países da América do Sul para o Canadá de 1980 a 2013')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

#Estrutura de repetição para deixar o eixo X de todos os gráficos de 5 em 5
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de Imigrantes')

#***Ajustando a escala dos valores***#
ymin = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)

plt.show()
```

Agora podemos ver que visualmente diferente dos gráficos anteriores, a colombia já parece ter uma densidade de pessoas imigrantes muito maior que os demais países analisados.

## Personalizando as visualizações

Aumentando fontes e adicionando novos elementos ao gráfico

```python
fig, ax = plt.subplots(figsize=(8,4))
#Aumentando a expessura da linha do gráfico com o lw e adicionando marcadores com o marker
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, marker='o')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
#Aumentando a fonte do título com fontsize e mudando o posicionamento do título com o loc
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=18, loc='left')
#Aumentando a fonte dos eixos com fontsize
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Numero de Imigrantes', fontsize=14)
#Aumentando o tamanho dos ticks do eixo x
ax.xaxis.set_tick_params(labelsize=12)
#Aumentando o tamanho dos ticks do eixo y
ax.yaxis.set_tick_params(labelsize=12)
#Adicionando um grid no grafico e deixando o estilo dele em dash com o linestyle
plt.grid(linestyle='--')

plt.show()
```

Adicionando cores, para cores mais especificas, consultar a documentação do matplotlib

```python
fig, ax = plt.subplots(figsize=(8,4))
#Mudando a cor da linha para verde com o parâmetro color
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, color='g')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', fontsize=18, loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Numero de Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

Mudando a cor de vários gráficos

```python
df.head()
```

```python
#pegando os países da america do sul
america_sul = df.query('Região == "América do Sul" ')
```

```python
america_sul
```

Ordenando dataframe em ordem crescente

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
#Lista de cores que vamos usar
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 
         'olive','navy', 'teal','tomato']

#criando gráfico de barras
fig, ax = plt.subplots(figsize = (12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de Imigrantes',fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

Destacando o Brasil no gráfico

```python
#Lista de cores que vamos usar
cores = []

#Destacando a cor do brasil
for pais in america_sul_sorted.index:
    if pais =='Brasil':
        cores.append('green')
    else:
        cores.append('silver')

#criando gráfico de barras
fig, ax = plt.subplots(figsize = (12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de Imigrantes',fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

fig.savefig('Imigracao_america_sul.png', transparent=False, dpi=300, bbox_inches='tight')
plt.show()
```

Adicionando anotações na figura

```python
cores = []

for pais in america_sul_sorted.index:
    if pais =='Brasil':
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize = (12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('Brasil foi o quarto país com mais imigrantes\npara o Canadá no período de 1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de Imigrantes',fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

#Adicionando o valor de cada barra no gráfico
for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

#Removendo a caixa em volta da figura
ax.set_frame_on(False)

#Retirando o eixo X
ax.get_xaxis().set_visible(False)

#Retirando os ticks do eixo y
ax.tick_params(axis='both', which='both', length=0)

plt.show()
```
