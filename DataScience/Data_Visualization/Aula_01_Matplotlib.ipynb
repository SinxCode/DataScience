```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Arrumando a Base

```python
df = pd.read_csv('C:/Users/gabrielbrito/DataScienceProjects/DataScience/Dados/imigrantes_canada.csv')
```

```python
df.head()   
```

```python
df.info()
```

```python
#Setando o index como o país
df.set_index('País', inplace=True)
```

```python
#Criando variável para armazenar os nomes das colunas de ano
anos = list(map(str,range(1980, 2014)))
```

```python
anos
```

```python
#Criando uma series somente com os dados do brasil e dos anos
brasil = df.loc['Brasil', anos]
```

```python
brasil
```

```python
#Criando dicionario da series anterior nomeando as colunas de ano e imigrantes
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}

#Transformando o dicionário em um df
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

## Plotagem

```python
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.xticks(['1980', '1985', '1990','1995', '2000', '2005','2010'])
plt.show()
```

## Criando uma Figura

A figura diferente de apenas usar o plot, permite mais funcionalidades e personalizações para o gráfico

```python
#Fig permite que uma figura seja criada (o espaço em branco), o ax representa os eixos
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
#Deixando o eixo X automaticamente em 5 em 5
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
#Setando o titulo em figura
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
#adicionando rotulos na figura
ax.set_xlabel('Ano')
ax.set_ylabel('Numero de Imigrantes')
plt.show()
```

## Plotando mais de um grafico na mesma figura

```python
#Dentro do parametro do subplot o 1 indica que ele tera uma linha só, e o 2 indica que serão 2 gráficos nessa msm linha
fig, axs = plt.subplots(1, 2, figsize=(15,5))
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot da Imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de Imigrantes')
axs[1].grid()

plt.show()
```
