```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/gabrielbrito/DataScienceProjects/DataScience/Dados/imigrantes_canada.csv')
```

```python
#Setando o index como o país
df.set_index('País', inplace=True)
```

```python
anos = list(map(str,range(1980, 2014)))
```

```python
#Criando um tema para que todos os demais gráficos tenham esse mesmo tema
sns.set_theme()
```

```python
#Criando um dataframe com os 10 paises com maior número de imigração
top_10 = df.sort_values('Total', ascending=False).head(10)
```

```python
top_10
```

Criando grafico de barras com o seaborn

```python
sns.barplot(data=top_10, x = top_10.index, y= 'Total')
```

Criando gráfico de barras horizontal

```python
sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')
```

Customizando o gráfico

```python
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

#Adicionando títulos
ax.set(title='Países com Maior Imigração para o Canadá\n1980 a 2013',
       xlabel='Número de Imigrantes',
       ylabel='')

plt.show()
```

Combinando o Seaborn com o MatplotLib para melhor personalização

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc = 'left', fontsize=16)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.set_ylabel('')

plt.show()
```

Paletas de Cores com o Seaborn

```python
#Criando função para gerar o gráfico com paleta de cores

def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=palette)
    ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=18)
    ax.set_xlabel('Número de Imigrantes', fontsize=14)
    ax.set_ylabel('')

    plt.show()
```

Paletas de cores sequenciais

```python
gerar_grafico_paleta('Blues')
```

```python
#Gerando o gráfico de modo que as cores mais escuras fiquem nos maiores valores
gerar_grafico_paleta('Blues_r')
```

```python
gerar_grafico_paleta('rocket')
```

Paleta de Extremidade

```python
gerar_grafico_paleta('coolwarm')
```

Paletas categóricas

```python
gerar_grafico_paleta('tab10')
```

Explorando temas

```python
sns.set_theme(style='dark')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='whitegrid')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='white')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='ticks')
gerar_grafico_paleta('tab10')
```

Retirando a caixinha do grafico

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, y= top_10.index, x='Total', orient='h', palette='tab10')
ax.set_title('Países com maior imigração para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.set_ylabel('')

#Tira linha superior e da direita
sns.despine()

plt.show()
```
